### Exercise 1: Simple Polynomial Plot

1. Plot a Polynomial Function:
   - Create an array `x` with 100 linearly spaced points between -10 and 10.
   - Create an array `y` for the polynomial function `y = 2x^3 - 5x^2 + 3x - 7`.
   - Plot `y` versus `x` with a blue line.
2. Customization:
   - Add labels to the x and y axes, and provide a title for the plot.
   - Change the figure size to 10x6 inches.

```python
import numpy as np

import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)

y = 2 * x**3 - 5 * x**2 + 3 * x - 7


plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='y = 2x^3 - 5x^2 + 3x - 7')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Function Plot')
plt.legend()

plt.show()
```

### Exercise 2: Exponential and Logarithmic Plot

1. Plot Exponential and Logarithmic Functions:
   - Create an array `x` with 500 linearly spaced points between 0.1 and 10.

   - Plot both functions on the same graph using different colors and line styles.
2. Customization:

```python
x_exp_log = np.linspace(0.1, 10, 500)

y1 = np.exp(x_exp_log)
y2 = np.log(x_exp_log)

plt.figure(figsize=(10, 6))
plt.plot(x_exp_log, y1, label='y = exp(x)', color='green', linestyle='-')
plt.plot(x_exp_log, y2, label='y = log(x)', color='red', linestyle='--')

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential and Logarithmic Functions')
plt.legend()

plt.savefig('exponential_logarithmic_plot.png', dpi=100)

plt.show()
```

### Exercise 3: Figure and Subplots

1. Create Subplots with Different Functions:
   - Create a figure with two subplots in one row.
   - In the first subplot, plot `y = tan(x)` and `y = arctan(x)` for `x` ranging from -2 to 2.
   - In the second subplot, plot `y = sinh(x)` and `y = cosh(x)` for `x` ranging from -2 to 2.
2. Customization:
   - Add labels to the axes, titles to the subplots, and legends.
   - Adjust the layout to avoid overlap.

```python

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

x_tan = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_tan = np.tan(x_tan)
y_arctan = np.arctan(x_tan)

axes[0].plot(x_tan, y_tan, label='y = tan(x)', color='blue')
axes[0].plot(x_tan, y_arctan, label='y = arctan(x)', color='orange')
axes[0].set_ylim(-10, 10)  
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = tan(x) and y = arctan(x)')
axes[0].legend()
axes[0].grid(True)

x_sinh_cosh = np.linspace(-2, 2, 500)
y_sinh = np.sinh(x_sinh_cosh)
y_cosh = np.cosh(x_sinh_cosh)

axes[1].plot(x_sinh_cosh, y_sinh, label='y = sinh(x)', color='green')
axes[1].plot(x_sinh_cosh, y_cosh, label='y = cosh(x)', color='red')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('y = sinh(x) and y = cosh(x)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### Exercise 7: 3D Surface Plot

1. Plot a 3D Surface Plot of a Mathematical Function:
   - Create arrays `X` and `Y` with values ranging from -5 to 5 with a step of 0.25.
   - Compute `Z = cos(sqrt(X^2 + Y^2))`.
   - Plot a 3D surface plot of `Z`.
2. Customization:
   - Change the colormap and the viewing angle of the plot. Add a color bar.

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.arange(-5, 5.25, 0.25)
y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2))


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_title('3D Surface Plot of Z = cos(sqrt(X^2 + Y^2))')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(elev=30, azim=45)  # Change the viewing angle
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)  # Add a color bar

plt.show()
```

### Exercise 16: 3D Scatter Plot

1. Plot a 3D Scatter Plot:
   - Create arrays `x`, `y`, and `z` each with 100 values drawn from a normal distribution.

2. Customization:

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

Z = np.exp(-0.1 * (X**2 + Y**2))

xpos, ypos = X.ravel(), Y.ravel()
zpos = np.zeros_like(xpos)
dx = dy = 0.5  
dz = Z.ravel()

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
bars = ax.bar3d(xpos, ypos, zpos, dx, dy, dz, shade=True, cmap='viridis')

ax.set_title('3D Bar Plot of Z = exp(-0.1 * (X^2 + Y^2))')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(elev=30, azim=45)  # Change the viewing angle

fig.colorbar(cm.ScalarMappable(cmap='viridis'), ax=ax, shrink=0.5, aspect=10)

# Show the plot
plt.show()
```

### Exercise 20: 3D Density Plot

1. Plot a 3D Density Plot:
   - Generate 1000 random points in 3D space.
   - Create a density plot to show where points are concentrated.
2. Customization:
   - Add labels to the axes. Adjust the transparency of the plot.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
z = np.random.normal(0, 1, 1000)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

hist, edges = np.histogramdd((x, y, z), bins=(20, 20, 20))
xpos, ypos, zpos = np.meshgrid(
    edges[0][:-1] + 0.5 * (edges[0][1] - edges[0][0]),
    edges[1][:-1] + 0.5 * (edges[1][1] - edges[1][0]),
    edges[2][:-1] + 0.5 * (edges[2][1] - edges[2][0]),
    indexing="ij"
)

xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = zpos.ravel()
values = hist.ravel()

max_value = values.max()
colors = plt.cm.viridis(values / max_value)
alphas = values / max_value
ax.bar3d(xpos, ypos, zpos, 0.1, 0.1, values, color=colors, alpha=0.7)

# Add labels to the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Density Plot')

# Show the plot
plt.show()
```
